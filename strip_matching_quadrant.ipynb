```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import stripROtools

#from matplotlib import style
#style.use('dark_background')
```

# UH DLC Micromegas

```python

# Mesh voltage values to include in analysis
UH_DLC_Vmesh = [620,630,640,650,660,670,680,690,700]

# Empty lists to fill with data
UH_DLC_xHyL_mu = []
UH_DLC_xHyL_mu_err = []
UH_DLC_xHyL_sigma = []
UH_DLC_xHyL_sigma_err = []

UH_DLC_xHyH_mu = []
UH_DLC_xHyH_mu_err = []
UH_DLC_xHyH_sigma = []
UH_DLC_xHyH_sigma_err = []

UH_DLC_xLyL_mu = []
UH_DLC_xLyL_mu_err = []
UH_DLC_xLyL_sigma = []
UH_DLC_xLyL_sigma_err = []

UH_DLC_xLyH_mu = []
UH_DLC_xLyH_mu_err = []
UH_DLC_xLyH_sigma = []
UH_DLC_xLyH_sigma_err = []

```

```python
for Vmesh in UH_DLC_Vmesh:


    print("--------------------UH DLC Micromegas, Vmesh = ", str(Vmesh),"--------------------")

    # Location of data
    file_loc = ["/Users/majdghrear/data/VMM3a_SRS/AUG23/UH_DLC/Fe55/"+str(Vmesh)+"Vmesh_bc_40p000_tac_60_ccs_4_cs_2_dt_200_mst_1_spc_1500_dp_200_cr_0p20-2p00_coin_center-of-mass_test.root"]

    # Create pandas data frame of the cluster info
    df_cluster = stripROtools.read_root(file_loc, clusters=True, hits=False)


    # Define additional columns
    df_cluster["nhits"]=df_cluster.apply(lambda row: len(row.strips0)+len(row.strips1) ,axis=1)
    # Calculate time offset based on time difference of max ADC hit in x and y
    df_cluster['maxADC_offset'] = df_cluster.apply(lambda row: row.times0[np.argmax(row.adcs0)] -  row.times1[np.argmax(row.adcs1)] , axis=1)


    # Fiducialize on single VMM in x and y
    df_fid = stripROtools.fiducializeQuadrant(df_cluster, 'xH', 'yL', min_hits=5)
    print("High x low y VMM combo")
    mu, mu_err, sigma, sigma_err = stripROtools.fitGauss(df_fid)

    UH_DLC_xHyL_mu += [mu]
    UH_DLC_xHyL_mu_err += [mu_err]
    UH_DLC_xHyL_sigma += [sigma]
    UH_DLC_xHyL_sigma_err += [sigma_err]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeQuadrant(df_cluster, 'xH', 'yH', min_hits=5)
    print("High x high y VMM combo")
    mu, mu_err, sigma, sigma_err = stripROtools.fitGauss(df_fid)
    # Print fit results
    print("Mu: ", mu, "+/-", mu_err )
    print("Sigma: ", sigma,"+/-", sigma_err  )

    UH_DLC_xHyH_mu += [mu]
    UH_DLC_xHyH_mu_err += [mu_err]
    UH_DLC_xHyH_sigma += [sigma]
    UH_DLC_xHyH_sigma_err += [sigma_err]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeQuadrant(df_cluster, 'xL', 'yL', min_hits=5)
    print("Low x low y VMM combo")
    mu, mu_err, sigma, sigma_err = stripROtools.fitGauss(df_fid)
    # Print fit results
    print("Mu: ", mu, "+/-", mu_err )
    print("Sigma: ", sigma,"+/-", sigma_err  )

    UH_DLC_xLyL_mu += [mu]
    UH_DLC_xLyL_mu_err += [mu_err]
    UH_DLC_xLyL_sigma += [sigma]
    UH_DLC_xLyL_sigma_err += [sigma_err]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeQuadrant(df_cluster, 'xL', 'yH', min_hits=5)
    print("Low x high y VMM combo")
    mu, mu_err, sigma, sigma_err = stripROtools.fitGauss(df_fid)
    # Print fit results
    print("Mu: ", mu, "+/-", mu_err )
    print("Sigma: ", sigma,"+/-", sigma_err  )

    UH_DLC_xLyH_mu += [mu]
    UH_DLC_xLyH_mu_err += [mu_err]
    UH_DLC_xLyH_sigma += [sigma]
    UH_DLC_xLyH_sigma_err += [sigma_err]

```

# Make plots for both detectors

```python
# Plot the mean time offset in all 4 quadrants

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2,figsize=(15,9),sharex='col', sharey='row')

ax1.errorbar(UH_DLC_Vmesh,UH_DLC_xLyH_mu,yerr=UH_DLC_xLyH_mu_err,fmt="*")
ax1.errorbar(UH_NoDLC_Vmesh,UH_NoDLC_xLyH_mu,yerr=UH_NoDLC_xLyH_mu_err,fmt="o")

ax2.errorbar(UH_DLC_Vmesh,UH_DLC_xHyH_mu,yerr=UH_DLC_xHyH_mu_err,fmt="*",label="DLC")
ax2.errorbar(UH_NoDLC_Vmesh,UH_NoDLC_xHyH_mu,yerr=UH_NoDLC_xHyH_mu_err,fmt="o",label="No DLC")
ax2.legend()

ax3.errorbar(UH_DLC_Vmesh,UH_DLC_xLyL_mu,yerr=UH_DLC_xLyL_mu_err,fmt="*")
ax3.errorbar(UH_NoDLC_Vmesh,UH_NoDLC_xLyL_mu,yerr=UH_NoDLC_xLyL_mu_err,fmt="o")

ax4.errorbar(UH_DLC_Vmesh,UH_DLC_xHyL_mu,yerr=UH_DLC_xHyL_mu_err,fmt="*")
DLC_xHyL_mean = np.mean(np.array(UH_DLC_xHyL_mu)[~np.isnan( np.array(UH_DLC_xHyL_mu))])
print("DLC xHyL mu = ", DLC_xHyL_mean)
ax4.axhline(DLC_xHyL_mean)
ax4.errorbar(UH_NoDLC_Vmesh,UH_NoDLC_xHyL_mu,yerr=UH_NoDLC_xHyL_mu_err,fmt="o")
NoDLC_xHyL_mean = np.mean(np.array(UH_NoDLC_xHyL_mu)[~np.isnan( np.array(UH_NoDLC_xHyL_mu))])
print("No DLC xHyL mu = ", NoDLC_xHyL_mean)
ax4.axhline(NoDLC_xHyL_mean, color = 'orange')

for ax in fig.get_axes():
    ax.set(xlabel='Mesh voltage [V]', ylabel='$\mu$ [ns]')

for ax in fig.get_axes():
    ax.label_outer()

fig.subplots_adjust(wspace=0, hspace=0)

```
