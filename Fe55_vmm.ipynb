```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import stripROtools
from stripROtools import colors

#from matplotlib import style
#style.use('dark_background')
```

```python
# Edges used by histograms
strip_edges = np.arange(-0.5,499.5,1.0)
adc_edges = np.arange(-0.5,1023.5,1.0)
```

# UH DLC Micromegas

```python

# Mesh voltage values to include in analysis
UH_DLC_Vmesh = [620,630,640,650,660,670,680,690,700]

# Empty lists to fill with data
UH_DLC_xLyL_gain = []
UH_DLC_xLyLgain_err = []
UH_DLC_xLyL_GR = []
UH_DLC_xLyL_GR_err = []
UH_DLC_xLyL_CS = []
UH_DLC_xLyL_mu_x = []
UH_DLC_xLyL_mu_y = []

UH_DLC_xLyH_gain = []
UH_DLC_xLyHgain_err = []
UH_DLC_xLyH_GR = []
UH_DLC_xLyH_GR_err = []
UH_DLC_xLyH_CS = []
UH_DLC_xLyH_mu_x = []
UH_DLC_xLyH_mu_y = []

UH_DLC_xHyL_gain = []
UH_DLC_xHyLgain_err = []
UH_DLC_xHyL_GR = []
UH_DLC_xHyL_GR_err = []
UH_DLC_xHyL_CS = []
UH_DLC_xHyL_mu_x = []
UH_DLC_xHyL_mu_y = []

UH_DLC_xHyH_gain = []
UH_DLC_xHyHgain_err = []
UH_DLC_xHyH_GR = []
UH_DLC_xHyH_GR_err = []
UH_DLC_xHyH_CS = []
UH_DLC_xHyH_mu_x = []
UH_DLC_xHyH_mu_y = []
```

```python
for Vmesh in UH_DLC_Vmesh:


    print("--------------------UH DLC Micromegas, Vmesh = ", str(Vmesh),"--------------------")

    # Location of data
    file_loc = ["/Users/majdghrear/data/VMM3a_SRS/AUG23/UH_DLC/Fe55/"+str(Vmesh)+"Vmesh_bc_40p000_tac_60_ccs_4_cs_2_dt_200_mst_1_spc_1500_dp_200_cr_0p00-1000p00_coin_center-of-mass_test.root"]

    # Create pandas data frame of the cluster and hit info
    df_cluster, df_hits = stripROtools.read_root(file_loc, clusters=True, hits=True)

    # Isolate x and y hits
    x_hits = df_hits.loc[df_hits.plane == 0].reset_index()
    y_hits = df_hits.loc[df_hits.plane == 1].reset_index()

    # Define additional columns
    # count number of hits in the event
    df_cluster["nhits"]=df_cluster.apply(lambda row: len(row.strips0)+len(row.strips1) ,axis=1)

    #compute number of electrons in event
    # Per Lucian 1 ADC ~ 1 mV
    df_cluster["electrons_x"] = df_cluster.adc0.apply(lambda x: 6240 * ( x  / 9.0 ) ) # 9 mV/fC is VMM gain setting for x channels, 170mV is the pedestal, 1200mV is th operating voltage 1024 is the number of possible ADC values
    df_cluster["electrons_y"] = df_cluster.adc1.apply(lambda x: 6240 * ( x / 4.5 ) ) # 4.5 mV/fC is VMM gain setting for y channels, 170mV is the pedestal, 1200mV is th operating voltage 1024 is the number of possible ADC values
    df_cluster["electrons"] = df_cluster.electrons_x + df_cluster.electrons_y
    df_cluster["gain"] = df_cluster["electrons"] / 167.5

    # Display raw y hits
    print("Raw y hits")
    plt.figure()
    n, bins, patches = plt.hist(y_hits.pos,bins=strip_edges,histtype='bar',color="goldenrod")
    n, bins, patches = plt.hist(y_hits.loc[y_hits.vmm == 13].pos,bins=strip_edges,histtype='bar',color="k")
    n, bins, patches = plt.hist(y_hits.loc[y_hits.vmm == 10].pos,bins=strip_edges,histtype='bar',color="k")
    plt.xlabel("strips y")
    plt.ylabel("count")
    plt.yscale("log")
    plt.tight_layout()
    plt.savefig("./figures/Fe55/UH_DLC/"+str(Vmesh)+"_raw_yhits.pdf")
    plt.show()


    # Display raw x hits
    print("Raw x hits")
    plt.figure()
    n, bins, patches = plt.hist(x_hits.pos,bins=strip_edges,histtype='bar',color="teal")
    n, bins, patches = plt.hist(x_hits.loc[x_hits.vmm == 2].pos,bins=strip_edges,histtype='bar',color="k")
    n, bins, patches = plt.hist(x_hits.loc[x_hits.vmm == 5].pos,bins=strip_edges,histtype='bar',color="k")
    plt.xlabel("strips x")
    plt.ylabel("count")
    plt.yscale("log")
    plt.tight_layout()
    plt.savefig("./figures/Fe55/UH_DLC/"+str(Vmesh)+"_raw_xhits.pdf")
    plt.show()

    # Visualize distribution of Fe55 clusters on readout plane
    print("Fe55 distribtuion on full readout plane")
    plt.figure()
    plt.hist2d(df_cluster.pos0, df_cluster.pos1, bins=[strip_edges,strip_edges], cmap=plt.cm.jet)
    plt.axvline(156,linestyle = "--", color="w")
    plt.axvline(217,linestyle = "--", color="w")
    plt.axvline(280,linestyle = "--", color="w")
    plt.axvline(342,linestyle = "--", color="w")
    plt.axhline(156,linestyle = "--", color="w")
    plt.axhline(217,linestyle = "--", color="w")
    plt.axhline(280,linestyle = "--", color="w")
    plt.axhline(342,linestyle = "--", color="w")
    plt.text(170.5,170.5,"(c)",color="w", fontsize="large", fontweight="bold")
    plt.text(295,295,"(b)",color="w", fontsize="large", fontweight="bold")
    plt.text(170.5,295,"(d)",color="w", fontsize="large", fontweight="bold")
    plt.text(295,170.5,"(a)",color="w", fontsize="large", fontweight="bold")
    plt.xlabel("strips x")
    plt.ylabel("strips y")
    plt.colorbar()
    plt.tight_layout()
    plt.savefig("./figures/Fe55/UH_DLC/"+str(Vmesh)+"_fe55_hits.pdf")
    plt.show()



    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=5, n_vmm_y=10, min_hits=5, map = "UH")
    print("Quadrant a CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)
    plt.tight_layout()
    plt.savefig("./figures/Fe55/UH_DLC/"+str(Vmesh)+"_CB_quad_a.pdf")


    UH_DLC_xHyL_gain += [mu]
    UH_DLC_xHyLgain_err += [mu_err]
    UH_DLC_xHyL_GR += [1.0*sigma/mu]
    UH_DLC_xHyL_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UH_DLC_xHyL_CS += [charge_sharing]
    UH_DLC_xHyL_mu_x += [mu_e_x]
    UH_DLC_xHyL_mu_y += [mu_e_y]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=5, n_vmm_y=13, min_hits=5, map = "UH")
    print("Quadrant b CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y  = stripROtools.fitCB(df_fid, plot=True)

    UH_DLC_xHyH_gain += [mu]
    UH_DLC_xHyHgain_err += [mu_err]
    UH_DLC_xHyH_GR += [1.0*sigma/mu]
    UH_DLC_xHyH_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UH_DLC_xHyH_CS += [charge_sharing]
    UH_DLC_xHyH_mu_x += [mu_e_x]
    UH_DLC_xHyH_mu_y += [mu_e_y]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=2, n_vmm_y=10, min_hits=5, map = "UH")
    print("Quadrant c CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y  = stripROtools.fitCB(df_fid, plot=True)

    UH_DLC_xLyL_gain += [mu]
    UH_DLC_xLyLgain_err += [mu_err]
    UH_DLC_xLyL_GR += [1.0*sigma/mu]
    UH_DLC_xLyL_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UH_DLC_xLyL_CS += [charge_sharing]
    UH_DLC_xLyL_mu_x += [mu_e_x]
    UH_DLC_xLyL_mu_y += [mu_e_y]

    # Fiducialize on single vmm in x and y
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=2, n_vmm_y=13, min_hits=5, map = "UH")
    print("Quadrant d CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)

    UH_DLC_xLyH_gain += [mu]
    UH_DLC_xLyHgain_err += [mu_err]
    UH_DLC_xLyH_GR += [1.0*sigma/mu]
    UH_DLC_xLyH_GR_err += [ ( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UH_DLC_xLyH_CS += [charge_sharing]
    UH_DLC_xLyH_mu_x += [mu_e_x]
    UH_DLC_xLyH_mu_y += [mu_e_y]



```

```python
for Vmesh in UoS_DLC_Vmesh:


    print("--------------------UoS Micromegas, Vmesh = ", str(Vmesh),"--------------------")

    # Location of data
    file_loc = ["/Users/majdghrear/data/VMM3a_SRS/AUG23/UoS/Fe55/"+str(Vmesh)+"Vmesh_bc_40p000_tac_60_ccs_4_cs_2_dt_200_mst_1_spc_1500_dp_200_cr_0p00-1000p00_coin_center-of-mass_test.root"]

    # Create pandas data frame of the cluster and hit info
    df_cluster, df_hits = stripROtools.read_root(file_loc, clusters=True, hits=True)

    # Isolate x and y hits
    x_hits = df_hits.loc[df_hits.plane == 0].reset_index()
    y_hits = df_hits.loc[df_hits.plane == 1].reset_index()

    # Define additional columns
    # count number of hits in the event
    df_cluster["nhits"]=df_cluster.apply(lambda row: len(row.strips0)+len(row.strips1) ,axis=1)

    #compute number of electrons in event
    # Per Lucian 1 ADC ~ 1 mV
    df_cluster["electrons_x"] = df_cluster.adc0.apply(lambda x: 6240 * ( x  / 1.0 ) ) # 1 mV/fC is VMM gain setting for x channels, 170mV is the pedestal, 1200mV is th operating voltage 1024 is the number of possible ADC values
    df_cluster["electrons_y"] = df_cluster.adc1.apply(lambda x: 6240 * ( x / 1.0 ) ) # 1 mV/fC is VMM gain setting for y channels, 170mV is the pedestal, 1200mV is th operating voltage 1024 is the number of possible ADC values
    df_cluster["electrons"] = df_cluster.electrons_x + df_cluster.electrons_y
    df_cluster["gain"] = df_cluster["electrons"] / 167.5

    # Display raw y hits
    print("Raw y hits")
    plt.figure()
    n, bins, patches = plt.hist(y_hits.pos,bins=strip_edges,histtype='bar',color="goldenrod")
    n, bins, patches = plt.hist(y_hits.loc[y_hits.vmm == 12].pos,bins=strip_edges,histtype='bar',color="k")
    n, bins, patches = plt.hist(y_hits.loc[y_hits.vmm == 10].pos,bins=strip_edges,histtype='bar',color="k")
    plt.xlabel("strips y")
    plt.ylabel("count")
    plt.yscale("log")
    plt.show()

    # Display raw x hits
    print("Raw x hits")
    plt.figure()
    n, bins, patches = plt.hist(x_hits.pos,bins=strip_edges,histtype='bar',color="teal")
    n, bins, patches = plt.hist(x_hits.loc[x_hits.vmm == 3].pos,bins=strip_edges,histtype='bar',color="k")
    n, bins, patches = plt.hist(x_hits.loc[x_hits.vmm == 2].pos,bins=strip_edges,histtype='bar',color="k")
    plt.xlabel("strips x")
    plt.ylabel("count")
    plt.yscale("log")
    plt.show()

    # Visualize distribution of Fe55 clusters on readout plane
    print("Fe55 distribtuion on full readout plane")
    plt.figure()
    plt.hist2d(df_cluster.pos0, df_cluster.pos1, bins=[strip_edges,strip_edges], cmap=plt.cm.jet)
    plt.axvline(122,linestyle = "--", color="w")
    plt.axvline(178,linestyle = "--", color="w")
    plt.axvline(179,linestyle = "--", color="w")
    plt.axvline(235,linestyle = "--", color="w")
    plt.axhline(64,linestyle = "--", color="w")
    plt.axhline(121,linestyle = "--", color="w")
    plt.axhline(179,linestyle = "--", color="w")
    plt.axhline(235,linestyle = "--", color="w")
    plt.text((122+178)/2.0 - 15,(64+121)/2.0 - 15,"(c)",color="w", fontsize="large", fontweight="bold")
    plt.text((179+235)/2.0 - 15,(179+235)/2.0 - 15,"(b)",color="w", fontsize="large", fontweight="bold")
    plt.text((122+178)/2.0 - 15,(179+235)/2.0 - 15,"(d)",color="w", fontsize="large", fontweight="bold")
    plt.text((179+235)/2.0 - 15,(64+121)/2.0 - 15,"(a)",color="w", fontsize="large", fontweight="bold")
    plt.xlabel("strips x")
    plt.ylabel("strips y")
    plt.colorbar()
    plt.show()


    # Fiducialize on single vmm in x and y
    # Quadrant a is xHyL
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=2, n_vmm_y=10, min_hits=5, map = 'UoS')
    print("Quadrant a CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)

    UoS_xHyL_gain += [mu]
    UoS_xHyLgain_err += [mu_err]
    UoS_xHyL_GR += [1.0*sigma/mu]
    UoS_xHyL_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UoS_xHyL_CS += [charge_sharing]
    UoS_xHyL_mu_x += [mu_e_x]
    UoS_xHyL_mu_y += [mu_e_y]

    # Fiducialize on single vmm in x and y
    # Quadrant b is xHyH
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=2, n_vmm_y=12, min_hits=5, map = 'UoS')
    print("Quadrant b CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)

    UoS_xHyH_gain += [mu]
    UoS_xHyHgain_err += [mu_err]
    UoS_xHyH_GR += [1.0*sigma/mu]
    UoS_xHyH_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UoS_xHyH_CS += [charge_sharing]
    UoS_xHyH_mu_x += [mu_e_x]
    UoS_xHyH_mu_y+= [mu_e_y]

    # Fiducialize on single vmm in x and y
    # Quadrant c is xLyL
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=3, n_vmm_y=10, min_hits=5, map = 'UoS')
    print("Quadrant c CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)

    UoS_xLyL_gain += [mu]
    UoS_xLyLgain_err += [mu_err]
    UoS_xLyL_GR += [1.0*sigma/mu]
    UoS_xLyL_GR_err += [( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UoS_xLyL_CS += [charge_sharing]
    UoS_xLyL_mu_x += [mu_e_x]
    UoS_xLyL_mu_y += [mu_e_y]


    # Fiducialize on single vmm in x and y
    # Quadrant d is xLyH
    df_fid = stripROtools.fiducializeVMM(df_cluster, n_vmm_x=3, n_vmm_y=12, min_hits=5, map = 'UoS')
    print("Quadrant d CrystalBall fit")
    # Fit Crystal Ball function
    beta, beta_err, m, m_err, mu, mu_err, sigma, sigma_err, charge_sharing, mu_e_x, mu_e_y = stripROtools.fitCB(df_fid, plot=True)

    UoS_xLyH_gain += [mu]
    UoS_xLyHgain_err += [mu_err]
    UoS_xLyH_GR += [1.0*sigma/mu]
    UoS_xLyH_GR_err += [ ( (sigma_err/sigma) + (mu_err/mu) ) * 1.0*sigma/mu ]
    UoS_xLyH_CS += [charge_sharing]
    UoS_xLyH_mu_x += [mu_e_x]
    UoS_xLyH_mu_y += [mu_e_y]


```
